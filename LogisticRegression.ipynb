```python
df['Target'] = df['Target'].map({'yes': 1, 'no': 0})
```

```python
x=df.iloc[:,:-1]
```

```python
x
```

```python
x=pd.get_dummies(df[[	'age','job','marital','education','default','balance','housing','loan','contact','day','month','duration','campaign','pdays','previous','poutcome']])
```

```python
x
```

```python
y=df.iloc[:,-1]
```

```python
y
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
model.fit(x,y)
```

```python
model.coef_
```

```python
y_p=model.predict(x)
```

```python
y_p
```

```python
Y_p_df=pd.DataFrame({'actual':y,'Predicted':y_p})
```

```python
Y_p_df
```

```python
from sklearn.metrics import confusion_matrix
c_m=confusion_matrix(y,y_p)
print(c_m)
```

```python
model.predict_proba(x)
```

```python
from sklearn.metrics import classification_report
print(classification_report)
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
fpr,tpr,thresholds=roc_curve(y,model.predict_proba(x)[:,1])


auc=roc_auc_score(y,y_p)
```

```python
plt.title('ROC Curve')
plt.plot(fpr,tpr,label='logit model',color='red')
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()

print('auc accuracy:',auc)
```
